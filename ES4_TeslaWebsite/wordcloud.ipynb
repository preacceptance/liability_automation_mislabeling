```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import spacy
import re
```

```python
# Read text for wordcloud
with open('wordCloud.txt', 'r') as f:
    text = f.read()

text = re.sub("driver", "driving", text)
text = re.sub("drives", "driving", text)
text = re.sub("drive", "driving", text)

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
text = ' '.join([token.text for token in doc if not token.is_stop])
text = text.lower()
```

```python
# Wordcloud settings
wordcloud = WordCloud(width=2000, height=1000, random_state=290, max_words=75, background_color="white", prefer_horizontal=.99, max_font_size=250, min_font_size=50)
# Process words and generate wordcloud
wc_dict = wordcloud.process_text(text)
wordcloud = wordcloud.generate_from_frequencies(wc_dict)

# Plot figure and save
plt.figure(figsize=(30, 16))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("wordcloud" + ".png", bbox_inches='tight')
```
